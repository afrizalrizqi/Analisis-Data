CLEANING DATA

cleaning data customers_df
- missing value : terdapat 18 missing values pada kolom gender
- duplicate data : terdapat 6 data duplikat
- inaccurate value : terdapat inaccurate value pada kolom age



```python
#menghilangkan data duplicate
customers_df.drop_duplicates(inplace=True)
customers_df.info() #mencari informasi pada data
customers_df.duplicated().sum() #data duplikasi yang awalnya 6 sekarang setelah di drop menjadi 0
```

```python
#menangani missing value
#ada 3 metode menangani missing value : dropping, imputation, interpolation
customers_df[customers_df.gender.isnull()] #cara ngecek untuk yang gendernya null
#karena berdasarkan output yang mengalami missing value ini masih banyak mengandung informasi penting maka sayang ketika ingin dibuang
#maka dengan ini menggunakan metode imputation
```

```python
#mengisi data yang missing dengan menggunakan value yang dominan sebagai pengganti missing value
#menggunakan fungsi value_counts()
customers_df.gender.value_counts() #ini menghitung dan mencari maka gender yang paling dominan
```

```python
#karena kita sudah tahu mana data yang dominan maka disini kita akan memasukkan data dominan tersebut untuk mengisi data yang missing
#dengan menggunakan fungsi .fillna()
customers_df.fillna(value="Prefer not to say",inplace=True)

customers_df.isnull().sum()
```

```python
#menangani inaccurate value
customers_df[customers_df.age == customers_df.age.max()]

#outputnya menghasilkan bahwa age yang dimasukkan yaitu 700 sehingga ini merupakan human error dengan memasukkan angka kelebihan 0 nya
```

```python
#sehingga kita bisa replace age nya dari 700 menjadi 70
customers_df.age.replace(customers_df.age.max(), 70, inplace=True)  #mereplace customer max dengan value 70
```

```python
customers_df[customers_df.age == customers_df.age.max()]
#disini terdapat human error lagi yaitu agenya 500 seharusnya dia 50 maka kita replace lagi
```

```python
delivery_time = orders_df["delivery_date"] - orders_df["order_date"] #untuk menghitung total secondsnya
delivery_time = delivery_time.apply(lambda x:x.total_seconds())
orders_df["delivery_time"] = round(delivery_time/86400) #86400 itu akan mengubah menjadi perhari

print(orders_df)
```

```python
orders_df.describe(include="all")
#untuk menampilkan data parameter statistik dari data orders_df
```

ANALISIS / EKSPLORASI DATA ORDERS_DF DAN CUSTOMERS_DF
membuat sebuah kolom baru bernama "status" pada data customers_df
Kolom tersebut memiliki nilai active jika pernah order setidaknya sekali
nilai non active jika belum pernah order sama sekali

```python
customer_id_in_orders_df = orders_df.customer_id.tolist()
customers_df["status"] = customers_df["customer_id"].apply(lambda x: "Active" if x in customer_id_in_orders_df else "Non Active")
customers_df.sample(5)
```

```python
customers_df.groupby(by="status").customer_id.count()
```

```python
orders_customer_df = pd.merge(
      left = orders_df,
      right = customers_df,
      how = "left",
      left_on = "customer_id",
      right_on = "customer_id"
)

orders_customer_df.head()
```

```python
#jumlah order berdasarkan kota
orders_customer_df.groupby(by="city").order_id.nunique().sort_values(ascending=False).reset_index().head(10)
```

```python
orders_customer_df.groupby(by="state").order_id.nunique().sort_values(ascending=False).reset_index().head(10)
```

```python
orders_customer_df.groupby(by="gender").order_id.nunique().sort_values(ascending=False)
```

```python
#jumlah order berdasarkan kelompok usia (youth, adult, senior)
orders_customer_df["age_group"] = orders_customer_df.age.apply(lambda x:"youth" if x<=24 else ("Senior" if x>64 else "Adults"))
orders_customer_df.groupby(by="age_group").order_id.nunique().sort_values(ascending=False)

#berdasarkan output didapatkan bahwa paling banyak itu adults sebanyak 681 orang
#kemudian diikuti oleh senior sebanyak 226 orang
#dan yang terakhir yaitu youth sebanyak 93 orang

#sehingga caranya kita buat kelompoknya dulu dengan menggunakan function apply(lambda x: .....)
```

EKSPLORASI DATA PRODUCT DAN SALES


```python
product_df.describe(include="all")
```

```python
#cara menampilkan data harga tertinggi, intinya jangan lupa ascending=False karena dia untuk mengurutkan dari yang terbesar ke terkecil
product_df.sort_values(by="price", ascending=False)
```

```python
product_df.groupby(by="product_type").agg({
    "product_id" : "nunique",
    "quantity" : "sum",
    "price" : ["min","max"]
})
```

```python
product_df.groupby(by="product_name").agg({
    "product_id" : "nunique",
    "quantity" : "sum",
    "price" : ["min","max"]
})
```

```python
sales_product_df = pd.merge(
    left = sales_df,
    right = product_df,
    how = "left",
    left_on = "product_id", #ini primary key dari tabel yang kiri
    right_on = "product_id" #ini merupakan key penghubungnya
)

sales_product_df.head()
```

```python
#melihat informasi penjualan berdasarkan tipe produk
sales_product_df.groupby(by="product_type").agg({
    "product_id" : "nunique",
    "quantity_x" : "sum",
    "total_price" : "sum"
})
```

```python
#melihat informasi penjualan berdasarkan nama produk
sales_product_df.groupby(by="product_name").agg({
    "sales_id": "nunique",
    "quantity_x": "sum",
    "total_price": "sum"
}).sort_values(by="total_price", ascending=False)
```

MENGEKSPLORASI SEMUA DATA UNTUK MELIHAT POLA PEMBELIAN BERDASARKAN DEMOGRAFI

```python
all_df = pd.merge(
    left = sales_product_df,
    right = orders_customer_df,
    how = "left",
    left_on = "order_id",
    right_on = "order_id"
)

all_df.head()
```

```python
#melihat preferensi pembelian berdasarkan state pelanggan dan tipe produk
all_df.groupby(by=["state","product_type"]).agg({
    "quantity_x" : "sum",
    "total_price" : "sum"
})
```

```python
#melihat selera tipe produk pelanggan berdasarkan gender dan kelompok usia
all_df.groupby(by=["gender","product_type"]).agg({  #munculin yang product_type nya itu disini bukan didalam agregatnya

    "quantity_x" : "sum",
    "total_price" : "sum"
})
```

```python
#untuk melihat tipe produk pelanggan berdasarkan kelompok usianya
all_df.groupby(by=["age_group","product_type"]).agg({
    "quantity_x" : "sum",
    "total_price" : "sum"
})
```
